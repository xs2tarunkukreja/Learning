```python
df = pd.read_csv("weather_data_5.csv", parse_dates=['day'])
type(df['day'][0]) 
```

```python
df.head(2)
```

```python
df.set_index('day', inplace=True)
df.head(2)
```

```python
df.fillna(0)
```

```python
df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': "no event"
})
```

```python
df.fillna(method="ffill") # forward fill. Previous value is filled.
```

```python
df.fillna(method="bfill")  # axis = "columns" - Left to Right..

# limit = 1 - Only direct consecutive copy. 2 consecutive Nan means one remain NaN.
```

```python
df.interpolate() # value between 28 and 32 will be 30 instead of 28 or 32.
# Linear Interpolate.
```

```python
df.interpolate(method="time") # It will be based on date.
```

```python
df.dropna() # row atleast 1 NaN
```

```python
df.dropna(how="all") # If all value are NaN
```

```python
df.dropna(thresh=1) # If there is any 1 not NaN then keep the row.
```

```python
# Bring Missing Dates
dt = pd.date_range("01-01-2017", "01-11-2017")
idx = pd.DatetimeIndex(dt)
df.reindex(idx)
```
